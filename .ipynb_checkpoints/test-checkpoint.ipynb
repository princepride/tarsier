```python
import re
def process_prompt(prompt, images_num: int = 8):
        if not images_num:
            prompt = prompt.replace("<image>", "").replace("<video>", "")
        elif images_num is not None:
            prompt = prompt.replace("<video>", "<image>"*images_num)
            image_token_num = len(re.findall('<image>', prompt, re.S))
            if image_token_num == 0:
                prompt_parts = re.findall(r'USER:(.*)ASSISTANT:(.*)', prompt, re.S)
                if prompt_parts and len(prompt_parts) == 2:
                    p1, p2 = prompt_parts
                else:
                    p1 = prompt
                    p2 = ''
                prompt = f"USER: {'<image>'*images_num + ' ' + p1.strip()} ASSISTANT: {p2.strip()}"
            assert image_token_num == images_num
        
        if not re.findall(r'USER:(.*)ASSISTANT:(.*)', prompt, re.S):
            prompt = f'USER: {prompt} ASSISTANT: '
        return prompt
```

```python
IGNORE_INDEX = -100
```

```python
import torch
s = processor(dataset['train']['text'][0]['prompt'], dataset['train']['video_file'][0], edit_prompt=True)
t = dataset['train']['text'][0]['response']
inputs = {k:v.to(model.device) for k,v in s.items() if v is not None}
source_id = inputs['input_ids']
target_id = processor.tokenizer.encode(t + '</s>', add_special_tokens=False, return_tensors='pt')[0]
input_id = torch.concat((source_id, target_id), dim=0)
label = input_id.clone()
label[:len(source_id[0])] = IGNORE_INDEX
```
